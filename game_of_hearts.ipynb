```
class Card:
  def __init__(self,typ,value):
    self.typ=typ
    self.value=value
```

Bring all the cards into the deck.

```
cards=[]
for i in range(1,14):
  cards.append(Card('heart',i))
for i in range(1,14):
  cards.append(Card('diamond',i))
for i in range(1,14):
  cards.append(Card('spade',i))
for i in range(1,14):
  cards.append(Card('club',i))
```

Shuffle and give cards to players

```
import numpy as np

player1_cards=[]
player2_cards=[]
player3_cards=[]
player4_cards=[]

for i in range(13):
  card_number=int(np.random.uniform(0,len(cards)))
  player1_cards.append(cards[card_number])
  cards.pop(card_number)

  card_number=int(np.random.uniform(0,len(cards)))
  player2_cards.append(cards[card_number])
  cards.pop(card_number)

  card_number=int(np.random.uniform(0,len(cards)))
  player3_cards.append(cards[card_number])
  cards.pop(card_number)

  card_number=int(np.random.uniform(0,len(cards)))
  player4_cards.append(cards[card_number])
  cards.pop(card_number)

print(cards)

print('Player 1:')
print()
for i in range(13):
  #print(i),
  print(player1_cards[i].typ+str(player1_cards[i].value)),

print('Player 2:')
print()
for i in range(13):
  #print(i),
  print(player2_cards[i].typ+str(player2_cards[i].value)),

print('Player 3:')
print()
for i in range(13):
  #print(i),
  print(player3_cards[i].typ+str(player3_cards[i].value)),

print('Player 4:')
print()
for i in range(13):
  #print(i),
  print(player4_cards[i].typ+str(player4_cards[i].value)),
```
